```python
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine, inspect
```

```python
# Create connection to database.
database = "music_data"
psql_engine = create_engine(f"postgresql://postgres:password@localhost/{database}")
```

```python
df = pd.read_csv("top_100_hip_hop_.csv")
```

```python
df.head()
```

```python
#  Check column types.
for col in df.columns:
    print(col, type(col[0]))
```

```python
# Pandas parses rank to int...
df[df['rank']>2]
```

```python
# Let's see what happens if we push this to Postgres if all columns are str.
df.to_sql(name="hip_hop",index=False,con=psql_engine,if_exists='replace')
```

```python
# Read in table.
query = "select * from hip_hop"
sql_table = pd.read_sql(query,psql_engine)
```

```python
# Looks good...
sql_table.head()
```

```python
# Still strings when read into pandas.
for col in sql_table.columns:
    print(col, type(col[0]))
```

```python
# Inspect sql columns. Looks good! Postgres parsed them!
inspector = inspect(psql_engine)
inspector.get_columns('hip_hop')
```
